## 1. Import libraries

```python
import urllib.request
```
This imports the `urllib.request` module, which is used to download files directly from the internet.

```python
url_address = r"https://gist.githubusercontent.com/aakashns/f6a004fa20c84fec53262f9a8bfee775/raw/f309558b1cf5103424cef58e2ecb8704dcd4d74c"
file_name = "italy-covid-daywise.csv"
```
The first line stores the dataset URL, and the second line stores the local filename that will be used after downloading the file.

```python
urllib.request.urlretrieve(url_address, file_name)
```
This downloads the CSV file from the given URL and saves it in the current working directory.


## 2. Load the dataset

```python
import pandas as pd
```
This imports Pandas and gives it the alias `pd`, which is the standard shorthand used in Python.

```python
covid_df = pd.read_csv(file_name)
```
This reads the downloaded CSV file and stores it inside a Pandas DataFrame named `covid_df`.

```python
print(covid_df)
print(type(covid_df))
```
The first line prints the full dataset, and the second line confirms that the object type is a DataFrame.


## 3. Explore the data

```python
covid_df.info()
```
This gives a quick summary of the DataFrame, including column names, non-null counts, and data types.

```python
print(covid_df.describe())
```
This shows descriptive statistics such as count, mean, standard deviation, minimum, maximum, and quartiles for numeric columns.

```python
print(covid_df.shape)
print(covid_df.columns)
```
The `shape` tells us the number of rows and columns, while `columns` lists all column names.


## 4. Access rows and columns

```python
print(covid_df["new_cases"][243])
```
This accesses the value in the `new_cases` column at index 243.

```python
print(covid_df.new_tests)
```
This accesses the `new_tests` column using dot notation.

```python
print(covid_df.loc[243:246, ["new_cases", "new_deaths"]])
print(covid_df.at[243, "new_cases"])
print(covid_df.loc[243:246])
```
`loc` is used for label-based selection of rows and columns, while `at` is used to fetch a single value quickly.

```python
print(covid_df.new_tests.first_valid_index())
```
This finds the first index where `new_tests` has a non-null value.

```python
print(covid_df.loc[109:112])
print(covid_df.sample(10))
print(covid_df.head(5))
print(covid_df.tail(5))
```
These lines help inspect specific rows, random rows, the first few rows, and the last few rows of the dataset.


## 5. Copy the DataFrame

```python
covid_df_copy = covid_df.copy()
print(covid_df_copy.loc[245:247])
```
This creates a separate copy of the original DataFrame. Any changes made to the copy will not affect the original DataFrame.


## 6. Compute totals and rates

```python
total_covid_case = covid_df.new_cases.sum()
total_covid_death = covid_df.new_deaths.sum()
```
These lines calculate the total number of new cases and total number of new deaths across the dataset.

```python
print(f"Total cases of covid in Italy={total_covid_case} and total deaths={total_covid_death}")
```
This prints the totals using an f-string.

```python
death_rate = total_covid_death / total_covid_case
print(f"The recorded covid death rate in Italy={round((death_rate),2)}")
```
This calculates the death rate and prints the rounded value.

```python
inital_tests = 935310
total_covid_tests = inital_tests + covid_df.new_tests.sum()
```
The first line stores the number of initial tests already conducted, and the second line adds them to the daily test totals in the dataset.

```python
covid_positive_rate = (total_covid_case / total_covid_tests)
print(f"The covid positive rate in Italy as per records={round(covid_positive_rate,2)}")
```
This calculates and prints the overall positivity rate.


## 7. Filter records

```python
high_covid_cases = covid_df.new_cases > 1500
print(high_covid_cases)
```
This creates a Boolean mask that marks rows where daily cases are greater than 1500.

```python
high_covid_cases_records = covid_df[high_covid_cases]
print(high_covid_cases_records)
print(high_covid_cases_records.shape)
```
This filters the DataFrame to keep only high-case days and then prints the result and its size.

```python
high_positive_rate_records = covid_df[covid_df.new_cases / covid_df.new_tests > covid_positive_rate]
print(high_positive_rate_records)
```
This selects records where the daily positivity rate is higher than the overall positivity rate.


## 8. Create and drop columns

```python
covid_df["positive_rate"] = covid_df.new_cases / covid_df.new_tests
print(covid_df)
```
This creates a new calculated column called `positive_rate`.

```python
covid_df = covid_df.drop(columns=["positive_rate"])
print(covid_df)
```
This removes the `positive_rate` column after using it.

```python
print(covid_df.sort_values("new_cases", ascending=False).head(10))
```
This sorts the DataFrame by `new_cases` in descending order and displays the top 10 rows.


## 9. Clean invalid data

```python
print(covid_df["new_cases"])
negative_records = covid_df[covid_df["new_cases"] < 0]
```
This prints the `new_cases` column and then filters the rows where new cases are negative.

```python
covid_df.at[172, "new_cases"] = (covid_df.at[171, "new_cases"] + covid_df.at[173, "new_cases"]) / 2
print(covid_df.at[172, "new_cases"])
```
This corrects the negative value by replacing it with the average of the previous and next rows.


## 10. Work with dates

```python
covid_df["date"] = pd.to_datetime(covid_df["date"])
print(covid_df.date)
```
This converts the `date` column into datetime format, which makes date-based analysis possible.

```python
covid_df["year"] = pd.DatetimeIndex(covid_df.date).year
covid_df["month"] = pd.DatetimeIndex(covid_df.date).month
covid_df["day"] = pd.DatetimeIndex(covid_df.date).day
covid_df["weekday"] = pd.DatetimeIndex(covid_df.date).weekday
covid_df["day_name"] = pd.DatetimeIndex(covid_df.date).day_name()
```
These lines extract useful date features such as year, month, day, weekday number, and weekday name.

```python
print(covid_df[50:60])
```
This prints rows 50 to 60 so the newly created date-related columns can be checked.


## 11. Month and weekday analysis

```python
print(covid_df[covid_df["month"] == 5])
```
This filters the DataFrame for records from May.

```python
print((covid_df[covid_df["month"] == 5])[["new_cases","new_tests","new_deaths"]].sum())
```
This calculates the total cases, tests, and deaths for May.

```python
avg_new_cases = (covid_df["new_cases"]).mean()
print(avg_new_cases)
```
This calculates the overall average number of new cases.

```python
avg_new_cases_sunday = (covid_df[covid_df["day_name"] == "Sunday"])["new_cases"].mean()
print(f"The average no. of cases on sunday is = {avg_new_cases_sunday} and overall average of new cases is = {avg_new_cases}")
```
This calculates the average number of new cases on Sundays and compares it with the overall average.


## 12. Group by month

```python
print((covid_df[(covid_df["year"] == 2020)].groupby("month"))[["new_cases", "new_deaths", "new_tests"]].sum().sort_values(by="new_deaths", ascending=False))
```
This groups the 2020 data by month and calculates the monthly totals, then sorts the output by deaths.

```python
print((covid_df[(covid_df["year"] == 2020)].groupby("month"))[["new_cases", "new_deaths", "new_tests"]].mean().sort_values(by="new_deaths", ascending=False))
```
This groups the 2020 data by month again, but this time calculates monthly average values.


## 13. Cumulative totals

```python
covid_df["total_cases"] = covid_df["new_cases"].cumsum()
covid_df["total_tests"] = covid_df["new_tests"].cumsum()
covid_df["total_deaths"] = covid_df["new_deaths"].cumsum()
print(covid_df[180:184])
```
These lines create running totals for cases, tests, and deaths using `cumsum()`.


## 14. Download and prepare location data

```python
url_address_2 = r"https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/locations.csv"
file_name_2 = "location.csv"
urllib.request.urlretrieve(url_address_2, file_name_2)
```
This downloads a second CSV file that contains country metadata such as population.

```python
location_df = pd.read_csv(file_name_2)
print(location_df.head(6))
```
This reads the location CSV and prints the first six rows.

```python
location_df.drop("location", axis=1, inplace=True)
print(location_df.head(6))
```
This removes the `location` column permanently from the DataFrame.

```python
location_df.rename(columns={"Country/Region":"country", "population_year":"census_year"}, inplace=True)
print(location_df.head(6))
```
This renames columns to make them cleaner and easier to merge.


## 15. Merge datasets

```python
covid_df["country"] = "Italy"
print(covid_df.tail(5))
```
This creates a `country` column in the COVID dataset so it can be used as a common merge key.

```python
merged_df = covid_df.merge(location_df, on="country")
print(merged_df.tail(6))
```
This merges the COVID DataFrame with the location DataFrame, similar to a SQL join.


## 16. Per-capita metrics

```python
merged_df["new_cases_per_mil"] = (merged_df["new_cases"] / merged_df["population"]) * (1000000)
merged_df["new_deaths_per_10mil"] = (merged_df["new_deaths"] / merged_df["population"]) * (10000000)
print(merged_df)
```
These calculations normalize daily cases and deaths based on population, which makes country-level analysis more meaningful.


## 17. Final result dataset

```python
italy_covid_result_df = merged_df[["date","new_cases","new_deaths","new_tests", "total_cases","total_deaths","new_cases_per_mil","new_deaths_per_10mil"]]
print(italy_covid_result_df)
```
This creates a final DataFrame with the selected columns needed for reporting and output.

```python
italy_covid_result_df.to_csv("Italy_covid_result", index=False)
```
This exports the final result to a CSV file without the index.
